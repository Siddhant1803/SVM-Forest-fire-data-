```python
#deleting month and day columns since we already have one hot encoded format of those features
df = raw.drop(['month','day'],axis=1)
df.head()
```

```python
# Custom Binary Encoding of Binary for size_category
df['size_category']=np.where(df['size_category'].str.contains("large"),1,0)
df
```

```python
df.columns
```

```python
#standardization of numerical data
from sklearn.preprocessing import StandardScaler
```

```python
le = StandardScaler()
df[['FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind', 'rain', 'area']]=le.fit_transform(df[['FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind', 'rain', 'area']])
```

```python
df
```

```python
#defining input data
X = df.iloc[:,:28]
```

```python
X.head(2)
```

```python
#defining target variable
Y = df['size_category']
Y.head(3)
```

```python
#splitting data using train_test_split
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.33,random_state=10,stratify=Y)
```

```python
#shape
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
#importing necessary library
from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report,confusion_matrix
```

Using GridSearhCV for hyper-parameter tunning

```python
clf = SVC()
param_grid=[{'kernel':['rbf','poly','linear'],
            'gamma':[50,5,10,0.5,20],
            'C':[15,14,13,12,11,10,0.1,0.001]}]
gsv = GridSearchCV(clf,param_grid,cv=10)
gsv.fit(x_train,y_train)
```

```python
gsv.best_params_, gsv.best_score_
```

```python
#instantiating an object
clf = SVC(kernel='linear',gamma=50,C=11)

#fitting on train data
clf.fit(x_train,y_train)
```

```python
#predecting on test data
y_pred=clf.predict(x_test)
```

```python
y_pred
```

```python
#predicting on whole input data
prediction = clf.predict(X)
```

```python
prediction
```

```python
confusion_matrix(y_test,y_pred)
```

```python
#confusion matrix of whole prediction
confusion_matrix(Y,prediction)
```

Inference - our model correctly predicting 376 small fire out of 378 and predicting 2 small fire as large fire i.e false predection. On the other hand 130 out of 139 are true prediction and 9 are false prediction.


```python
#classification report 
print(classification_report(y_test,y_pred))
```

```python
#classification report of whole prediction on input data
print(classification_report(Y,prediction))
```

for prediction of whole data, Accuracy score is 98%.
